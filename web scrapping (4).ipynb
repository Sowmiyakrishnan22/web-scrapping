## Scrapping website

```python
 url ="https://www.flipkart.com/audio-speaker-store?s_kwcid=AL!739!3!582822043916!e!!g!!flipkart&gclsrc=aw.ds&&semcmpid=sem_8024046704_sitelink_audio_goog&gclid=Cj0KCQjwzLCVBhD3ARIsAPKYTcT5OsE-b2WAju3aL85vubhNyaXaDinGDB_6CrveR2duoNGDVSmaGmAaAuGwEALw_wcB"
```

## website response

```python
res = requests.get(url)
```

```python
res
```

## parsing

```python
soup = BeautifulSoup(res.text)
```

```python
soup
```

## speaker

```python
speaker=soup.select(".s1Q9rs")
```

```python
speaker
```

```python
prices = soup.select('._30jeq3')
```

```python
prices
```

```python
rating=soup.select("._3LWZlK")
```

```python
rating
```

```python
speaker_names=[]
```

```python
for item in speaker:
    speaker_names.append(item.text)
```

```python
speaker_names
```

```python
price_title=[]
```

```python
for item in prices:
    price_title.append(item.text)
```

```python
price_title
```

```python
rating_title=[]
```

```python
for item in rating:
    rating_title.append(item.text)
```

```python
rating_title
```

## Data Frame

```python
df=ss.DataFrame(speaker_names,columns=['speaker name'])
```

```python
df
```

```python
df['price_title']=price_title
```

```python
df
```

```python
df["raing_title"]=rating_title
```

```python
df
```

```python
df.to_csv('sa.csv')
```
